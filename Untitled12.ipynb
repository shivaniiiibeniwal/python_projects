```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dairy=pd.read_csv("dairy_dataset.csv")
dairy.info()
```

```python
# 1.What is the average expiry days of each product.
gb=dairy.groupby(["Product Name","Storage Condition"]).agg({"Shelf Life (days)":"mean"})
gb
```

```python
# 2.What is the average shelf life for different storage conditions.
gb1=dairy.groupby("Storage Condition").agg({"Shelf Life (days)":"mean"})
gb1
```

```python
# 3.Mention the name of the products with maximum and minimum shelf life.
gb2=dairy.groupby("Product Name").agg({"Shelf Life (days)":"mean"})
gb2

max_storage=gb2.idxmax()
print("Maximum Storage Product is:",max_storage)

min_storage=gb2.idxmin()
print("Minimum Storage Product is:",min_storage)
```

```python
gb5=dairy.groupby(["Product Name", "Storage Condition"]).agg({"Shelf Life (days)":"mean"})
gb5
```

```python
# 4.Summarize the data on the basis of brands and land acquired by them,cows and the revenue generated by them.
GB=dairy.groupby("Brand").agg({"Total Land Area (acres)":"sum","Number of Cows":"sum","Approx. Total Revenue(INR)":"sum"})
GB
```

```python
# 5.What is the overall revenue generated by all brands.
total_revenue=dairy["Approx. Total Revenue(INR)"].sum()
print("Total Revenue by all brands:",total_revenue)
```

```python
# 6.Which brand has the maximum total land acquired?
gb4=dairy.groupby("Brand").agg({"Total Land Area (acres)":"sum"}).idxmax()
print("Brand with maximum land acquired:",gb4)
```

```python
# 7.How does the distribution of number of cows vary across brands?
sns.barplot(x="Brand",y="Number of Cows",data=dairy,palette="rocket")
plt.title("Distribution of Number of Cows Across Brands")
plt.xticks(x="Brand",rotation=45)
plt.show()
```

```python
# 8.Which product has the highest sales in each brand?
max_sales_product_per_brand = dairy.groupby('Brand')['Product Name'].apply(lambda x: x[dairy.loc[x.index, 'Approx. Total Revenue(INR)'].idxmax()])
print("Product with the highest sales in each brand:\n", max_sales_product_per_brand)
```

```python
# 9.Does number of cows has any dependency on area acquired.
gb_area=dairy.groupby("Farm Size").agg({"Number of Cows":"mean"})
gb_area
# Land area has no such dependency on area aquired
```
